## Tidying Data
In this stage, we take our newly formed dataframe and clean it up a little. We are lucky that the creator of the csv file left us with a resonably tidy data table to begin, but there are a lot of sloppily organized data sets out there. The key things to be weary of when tidying data are:

* Making sure variable name and not values are stored in column headers
* Repetition of data is minimized
* Column headers are clearly and appropriately named
* Unnecessary data is removed
* Unavailable data is dealt with

While these are just a couple of the core principles to cleaning data, there are many ways in which you can organize your data to allow not only the reader a clear view of the table but also yourself! All in all the tidying data helps speed along the path to quicker and more thorough analysis. 

For our dataframe, I chose to first remove the redundant columns from the dataframe. In this case, I removed the columns that contained the regional sales. One always has to be careful when removing data not to affect the results in any way. Here, I thought it was appropriate to remove these columns since we can simply consider the global sales as an aggragate.

```python
# Removing Redundant Columns
videogame_data = videogame_data.drop(['NA_Sales','EU_Sales','JP_Sales','Other_Sales'], axis=1)

videogame_data.head()
```

Next up is to simply rename and rearrange the columns. I rearrange the columns based on two factors. Firstly, the relatable variables should be neighboring. This helps reduce confusion when analyzing. Secondly, the more "pressing"/"important" variable towards the beginning of the table and less "pressing"/"important" variables towards the back. This is more a matter of preference, but personally, I find that organizing based on what variables I think matter helps with analyzation down the road. 

For our data frame, there are certainly variable names that could be clearer as well as a drastic need to reorganize columns. We shall tidy below:

```python
# Renaming the columns
videogame_data.columns = ['Videogame','Platform','Release Year','Genre','Publisher','Global Sales','Critic Score','Critic Count','User Score','User Count','Developer','Rating']

# Organizing Columns
videogame_data = videogame_data[['Videogame','Global Sales','Developer','Publisher','Platform','Genre','Rating','Release Year','Critic Score','Critic Count','User Score','User Count']]

videogame_data.head()
```

There our data is looking much better! A tidy data frame goes a long way towards a cleaner analysis. There is only one glaring mess left to tidy. What do we do with all the unavailable values (marked by NaN by pandas)? 
## Missing Data
When dealing with missing data, we always have to ask why the data is missing in order to properly deal with the missing values. Rarely is it that data is missing in a completely random fashion. On Kaggle, where we retrieved our data set, the explanation given is that Metacritic, where the data was scraped from, only covers a subset of the platforms. Another explanation given to us is that a portion of the games did not have information for all the variables. Metacritic more consistently reviews more popular games, but ultamately leaves the decision of game inclusion is based on whether the game was reviewed by multiple [publications](www.metacritic.com/faq#item30). Now that we know metacritic is an aggragation of reviews from other publications, we are more confident that explicit bias in our data is minimal. 
For sake of simplicity for this tutorial we are going to simply remove the rows that contain NaN. However there are [more ways](https://measuringu.com/handle-missing-data/) to deal with missing data involving various machine learning techniques.
#### Important Note*
dropping values is not always as clean. Without asking why the data points are missing, one can possible remove a subcategory of data that is pertinent to the analysis! You might be inserting an unforeseen bias into your analysis as well. Always ask why the data is missing!

```python
# Removing the missing values marked by NaN
videogame_data = videogame_data.dropna()
videogame_data.head()
```

```python
# Checking how many values we are left
videogame_data.shape
```

Our data frame is now tidied up and the missing values have been dealt with. As one can see, there are still roughly 6900 data points leftover to work with. With our data cleaned we can now properly begin our analysis. Finally!

## Analysis
The easiest way to discern whether two numeric variables share a relationship is to fit a linear regression to the two variables. This is where the Seaborn tool comes in. While pandas, through numpy, does have some great plotting options, I personally found Seaborn to be an simpler alternative to fitting linear regression. Seaborn is built on top of the matplotlib library and offers a wider variety of plotting options geared towards those focusing on statistical analysis. Click here for the proper documentation on [Seaborn](https://seaborn.pydata.org/) and on [Matplotlib](https://matplotlib.org/). 

First, we will see what the relationship between global sales and critic scores looks like. Using an educated guess, I would suspect that there is a positive relationship between higher critic scores and higher global sales because, intuitively, games with better ratings should have a larger appeal. 

```python
# First, I am going to rescale the critic score to align with the scaling on the user score for ease of interpretation.
videogame_data[['Critic Score']] = videogame_data[['Critic Score']].astype('int')
videogame_data[['Critic Score']] = videogame_data[['Critic Score']].apply(lambda x: x/10)

# Now, plotting the data
ax = sns.regplot(x='Global Sales', y='Critic Score', data = videogame_data)
ax.set(xlabel = 'Global Sales (by million)', ylabel = 'Critic Score (ranged 0-10)')
ax.set_title('Critic Ratings vs Global Sales')
plt.ylim(0, 11)
plt.xlim(0,26)
plt.show()
```
