## Extracting the Data:
We have our libraries installed and imported and our data on hand. Now what? This is where our first two libraries, pandas and numpy come in. With these two libraries in cooperation, we are provided an effecive tool for data analysis. We begin by creating a pandas dataframe and storing the data (formatted in a CSV (comma seperated values file)) into this dataframe.

For a more detailed look on pandas and numpy visit [pandas quickstart](http://www.gregreda.com/2013/10/26/intro-to-pandas-data-structures/) and [numpy quickstart](https://www.python-course.eu/numpy.php)

```python
# Loading the data
videogame_data = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")

# Displaying the newly created dataframe
videogame_data.head()
```

Here we have a look at what our data looks like stored in the dataframe. All the information from the raw CSV file should be stored into this dataframe. The first step towards your life as a data scientist is done! Don't get too excited, because extracting the data is always the easiest step. Onto the next!

## Tidying Data
In this stage, we take our newly formed dataframe and clean it up a little. We are lucky that the creator of the csv file left us with a resonably tidy data table to begin, but there are a lot of sloppily organized data sets out there. The key things to be weary of when tidying data are:

* Making sure variable name and not values are stored in column headers
* Repetition of data is minimized
* Column headers are clearly and appropriately named
* Unnecessary data is removed
* Unavailable data is dealt with

While these are just a couple of the core principles to cleaning data, there are many ways in which you can organize your data to allow not only the reader a clear view of the table but also yourself! All in all the tidying data helps speed along the path to quicker and more thorough analysis. 

For our dataframe, I chose to first remove the redundant columns from the dataframe. In this case, I removed the columns that contained the regional sales. One always has to be careful when removing data not to affect the results in any way. Here, I thought it was appropriate to remove these columns since we can simply consider the global sales as an aggragate.

```python
# Removing Redundant Columns
videogame_data = videogame_data.drop(['NA_Sales','EU_Sales','JP_Sales','Other_Sales'], axis=1)

videogame_data.head()
```

Next up is to simply rename and rearrange the columns. I rearrange the columns based on two factors. Firstly, the relatable variables should be neighboring. This helps reduce confusion when analyzing. Secondly, the more "pressing"/"important" variable towards the beginning of the table and less "pressing"/"important" variables towards the back. This is more a matter of preference, but personally, I find that organizing based on what variables I think matter helps with analyzation down the road. 

For our data frame, there are certainly variable names that could be clearer as well as a drastic need to reorganize columns. We shall tidy below:

```python
# Renaming the columns
videogame_data.columns = ['Videogame','Platform','Release Year','Genre','Publisher','Global Sales','Critic Score','Critic Count','User Score','User Count','Developer','Rating']

# Organizing Columns
videogame_data = videogame_data[['Videogame','Global Sales','Developer','Publisher','Platform','Genre','Rating','Release Year','Critic Score','Critic Count','User Score','User Count']]

videogame_data.head()
```

There our data is looking much better! A tidy data frame goes a long way towards a cleaner analysis. There is only one glaring mess left to tidy. What do we do with all the unavailable values (marked by NaN by pandas)? 
## Missing Data
When dealing with missing data, we always have to ask why the data is missing in order to properly deal with the missing values. Rarely is it that data is missing in a completely random fashion. On Kaggle, where we retrieved our data set, the explanation given is that Metacritic, where the data was scraped from, only covers a subset of the platforms. Another explanation given to us is that a portion of the games did not have information for all the variables. Metacritic more consistently reviews more popular games, but ultamately leaves the decision of game inclusion is based on whether the game was reviewed by multiple [publications](www.metacritic.com/faq#item30). Now that we know metacritic is an aggragation of reviews from other publications, we are more confident that explicit bias in our data is minimal. 
For sake of simplicity for this tutorial we are going to simply remove the rows that contain NaN. However there are [more ways](https://measuringu.com/handle-missing-data/) to deal with missing data involving various machine learning techniques.
#### Important Note*
dropping values is not always as clean. Without asking why the data points are missing, one can possible remove a subcategory of data that is pertinent to the analysis! You might be inserting an unforeseen bias into your analysis as well. Always ask why the data is missing!

```python
# Removing the missing values marked by NaN
videogame_data = videogame_data.dropna()
videogame_data.head()
```

```python
# Checking how many values we are left
videogame_data.shape
```

Our data frame is now tidied up and the missing values have been dealt with. As one can see, there are still roughly 6900 data points leftover to work with. With our data cleaned we can now properly begin our analysis. Finally!

## Analysis
The easiest way to discern whether two numeric variables share a relationship is to fit a linear regression to the two variables. This is where the Seaborn tool comes in. While pandas, through numpy, does have some great plotting options, I personally found Seaborn to be an simpler alternative to fitting linear regression. Seaborn is built on top of the matplotlib library and offers a wider variety of plotting options geared towards those focusing on statistical analysis. Click here for the proper documentation on [Seaborn](https://seaborn.pydata.org/) and on [Matplotlib](https://matplotlib.org/). 

First, we will see what the relationship between global sales and critic scores looks like. Using an educated guess, I would suspect that there is a positive relationship between higher critic scores and higher global sales because, intuitively, games with better ratings should have a larger appeal. 

```python
# First, I am going to rescale the critic score for ease of interpretation.
videogame_data[['Critic Score']] = videogame_data[['Critic Score']].astype('int')
videogame_data[['Critic Score']] = videogame_data[['Critic Score']].apply(lambda x: x/10)

# Now, plotting the data
ax = sns.regplot(x='Global Sales', y='Critic Score', data = videogame_data)
ax.set(xlabel = 'Global Sales (by million)', ylabel = 'Critic Score (ranged 0-10)')
ax.set_title('Critic Ratings vs Global Sales')
plt.ylim(0, 11)
plt.xlim(0,26)
plt.show()
```

As we expected, there is a positive relationship between the critic ratings and the total sales of a videogame. This is shown through the positive linear regression that we plotted through the scatter graph. We zoomed in on the data in order to properly see the linear regression, so note that while we cannot see beyond 25 million global copies sold, they were accounted for when plotting this data. 
Lets continue by doing the same thing for the user score.

```python
# First, I am going to rescale the user score for ease of interpretation.
videogame_data[['User Score']] = videogame_data[['User Score']].astype('float')

# Now, plotting the data
ax = sns.regplot(x='Global Sales', y='User Score', data = videogame_data)
ax.set(xlabel = 'Global Sales (by million)', ylabel = 'User Score (ranged 0-10)')
ax.set_title('User Ratings vs Global Sales')
plt.ylim(0, 11)
plt.xlim(0, 25)
plt.show()
```

Again, we find a positive relationship, this time between the user ratings and the total sales of a videogame. The linear regression has a slightly less steep linear regression. This implies that the relationship between user ratings and global sales is less important than critic ratings when deciding if a video game sells. These are factors to consider when arriving at the machine learning stage. This is likely because the average person will trust a veteran gamer or in this case a critic, more than they will trust the average person. Another reason might be because the average person is more reluctant to dole out a bad review than a critic might be. In either case, we now will prioritize critic reviews over user reviews.

The first thing to do is see how many different categories there are within a qualitative variable. For example, there are probably lots of different types of publishers as well as developers. If there are too many categories, then any plots we decide to make will have the points distributed too widely between the hundereds of categories.

```python
# Finding the number of unique values within the specified categories
pub_count = videogame_data.Publisher.nunique()
dev_count = videogame_data.Developer.nunique()
plat_count = videogame_data.Platform.nunique()
rate_count = videogame_data.Rating.nunique()
genre_count = videogame_data.Genre.nunique()

# printing the values
print("Publisher count: " + str(pub_count))
print("Developer count: " + str(dev_count))
print("Platform count: " + str(plat_count))
print("Rating count: " + str(rate_count))
print("Genre count " + str(genre_count))
```

Thus, we can see that the publisher and developer have a lot of different values within them. These categories are far too large to properly plot in any meaningful way. However, the Platform, Rating, and Genre categories have far fewer unique values. For simplicity's sake, I will go over only the Rating and Genre category.

To anaylze qualitative factor in a data set, there are many different types of plots to choose from: box plot, violin plot, swarm plot, etc. We will go over violin plots and strip plots in this guide. Using a violin plot, I will plot ratings over global sales. A violin plot will show the distribution of sales across each rating.  

```python
# Plotting a violin plot of ratings over global sales
ax = sns.violinplot(x = 'Rating',y = 'Global Sales', data = videogame_data)
plt.title("Ratings vs Global Sales")
plt.show()
```

```python
# Plotting again with the intention of zooming in on the bulk of the densities
ax = sns.violinplot(x = 'Rating',y = 'Global Sales', data = videogame_data)
plt.title("Ratings vs Global Sales")
plt.ylim(0, 5)
plt.show()
```

Three things of note in this plot. The first is that there are the ratings that only have 1 data point a piece to plot, which serves primarily to waste precious space in the graph. Secondly, The games that sell more than ~13 million copies are very sparse as we can see from the densities of the ratings north of 13 million. Thirdly, and most importantly, the densities are all very similar in range and spread between all the prevalent categories. This indicates that ratings do not factor into whether a video game sells. This makes sense as, theoretically, video games are created for all audiences and children to adults should all be invested in videogames semi-equally. 

Now, I will plot genre vs. global sales using a strip plot. A strip plot is useful for visualizing data that has the extreme data points that we saw in the ratings violin plot. A strip plot helps more in this situation because there are twice as many different genres than ratings.

```python
# Plotting the strip plot of genres vs global sales
ax = sns.stripplot(x = 'Genre', y = 'Global Sales', data = videogame_data, jitter = True, size = 3.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.title("Genre vs Global Sales")
plt.show()
```

```python
# Zooming in on the strip plot
ax = sns.stripplot(x = 'Genre', y = 'Global Sales', data = videogame_data, jitter = True, size = 3.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.title("Genre vs Global Sales")
plt.ylim(0,14)
plt.show()
```

Here we can see that again the densites do not vary much between genres. There are a couple categories that tend to do less well than the standards such as puzzles, fighting, role-playing, adventure, and strategy. But typically, we can see that the genres do not differ in densities. Thus, we can see that genre is not a good predictor for whether a game will sell or not. This makes sense as the classifications of these genres are probably aggregated from smaller subacategories. Thus, we get categories that are evenly layered as to even out the global sales distribution.

Now that all the variables we selected have been analyzed, we know that the critic score has a strong positive correlation for global sales. The last step is to perform tests that verify and measure the strength of the relationship between the found variables: in this case, critic scores vs global sales.

Below we will perform an f-test to see the strength of the relationship between critic score and global sales and user score and global sales indeed favors the critic score model. An f-test uses the OLS, or ordinary least squares method, which we unwittingly utilized when plotted linear regression on our data. More reading on the topic can be found [here](http://blog.minitab.com/blog/adventures-in-statistics-2/understanding-analysis-of-variance-anova-and-the-f-test). 

```python
# Compute the f value first by calculating the ratio of variances between our two variables.
f = (np.var(videogame_data['Critic Score']))/(np.var(videogame_data['Global Sales']))

# Computing f, distributed by the lengths - 1
df1 = len(videogame_data['Critic Score'])
df2 = len(videogame_data['Global Sales'])

# Computing the f-test p-value.
p_value = scipy.stats.f.cdf(f, df1, df2)
p_value
```

```python
# Compute the f value first by calculating the ratio of variances between our two variables.
f = (np.var(videogame_data['User Score']))/(np.var(videogame_data['Global Sales']))

# Computing f, distributed by the lengths - 1
df1 = len(videogame_data['Critic Score'])
df2 = len(videogame_data['Global Sales'])

# Computing the f-test p-value.
p_value = scipy.stats.f.cdf(f, df1, df2)
p_value
```

The higher the p-value, the more inaccurate the model is. Thus, it indeed holds that the critic score is indeed the better model.
